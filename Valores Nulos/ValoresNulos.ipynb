```python
import pandas as pd 
import missingno as msno 
```

```python
#Leemos el archivo 
df = pd.read_csv('ventas_totales.csv')
df.head()

```

Usamos la herramiena vista con el profesor Fransisco para poder ver cuales son los datos que nos faltan

```python
msno.matrix(df) #Matriz de nulos
```

Podemos ver que tenemos valores faltantes en las columnas de salon_ventas, tarjetas_debito, tarjetas_credito, otros_medios, subtotal_ventas_alimentos_bebidas, bebidas, almacen, panaderia, lacteos, carnes, verdureria_fruteria, alimentos_preparados_rostiseria, indumentaria_calzado_textiles_hogar, electronicos_articulos_hogar y otros.

Para poder decidir como tratar los datos faltantes, tenemos que entender un poco de que tratan las columnas.

```python
# Para tratar los nulos en la columna de salon_ventas, vamos a tomar el valor de ventas_totales_canal_venta, debido a que los valores son muy similares y se refieren a una misma transacción. 

df['salon_ventas'] = df['salon_ventas'].fillna(df['ventas_totales_canal_venta'])
```
